```python
# 1. Преобразование столбца 'Date' в формат даты и времени (datetime)
# Это необходимо для корректной работы с датами, например, для сравнения, сортировки и нахождения медианы.
df['Date'] = pd.to_datetime(df['Date'])

# 2. Нахождение медианной даты в столбце 'Date'
# Медиана - это значение, которое делит упорядоченный набор данных на две равные части.
# В данном случае, middle_date будет содержать дату, которая делит все даты в столбце 'Date' пополам.
middle_date = df['Date'].quantile(0.5)

# 3. Вывод значения middle_date на экран
# Это позволит нам увидеть, какая дата была найдена в качестве медианы.
print(f"Медианная дата: {middle_date}")
```

**Медианная дата была найдена с целью разделения временного ряда данных о погоде на два равных периода.** Это позволило сравнить распределения температурных значений (MinTemp и MaxTemp) до и после центральной точки наблюдений и выявить потенциальные различия в динамике температур.

Для сравнения распределений температурных значений (MinTemp) в двух выделенных периодах, данные были преобразованы в векторы. Затем было вычислено косинусное сходство между этими векторами, которое отражает степень их схожести. Дополнительно, был вычислен угол между векторами, предоставляющий наглядное представление о различиях в распределениях температур.

```python
# Разделение данных на два периода, используя медианную дату (middle_date)
# A_values будет содержать значения 'MinTemp' для дат ПОСЛЕ middle_date.
A_values = df[df['Date'] > middle_date]['MinTemp']
# B_values будет содержать значения 'MinTemp' для дат ДО или РАВНЫХ middle_date.
B_values = df[df['Date'] <= middle_date]['MinTemp']

# Преобразование данных в массивы NumPy для эффективных вычислений
# Преобразуем A_values и B_values (Series из pandas) в массивы NumPy (A и B).
A = np.array(A_values.tolist())
B = np.array(B_values.tolist())

# Приведение векторов A и B к одинаковой длине
# Находим минимальную длину между A и B.
min_len = min(len(A), len(B))
# Обрезаем оба вектора до минимальной длины, чтобы они были совместимы для вычисления косинусного сходства.
A = A[:min_len]
B = B[:min_len]

# Вычисление косинусного сходства между векторами A и B
# Скалярное произведение векторов A и B.
dot_product = np.dot(A, B)
# Норма (длина) вектора A.
norm_A = np.linalg.norm(A)
# Норма (длина) вектора B.
norm_B = np.linalg.norm(B)
# Косинусное сходство рассчитывается как скалярное произведение, деленное на произведение норм векторов.
cosine_similarity = dot_product / (norm_A * norm_B)

# Вывод результата - значения косинусного сходства
print(f"Косинусное сходство: {cosine_similarity}")
# Вычисление угла сходства в градусах
angle_radians = np.arccos(cosine_similarity) # Вычисляем угол в радианах
angle_degrees = np.degrees(angle_radians)    # Переводим угол в градусы

print(f"Угол сходства: {angle_degrees:.2f} градусов")
```

```python
# Угол между двумя векторами в градусах
angle_deg = 21.54
angle_rad = np.deg2rad(angle_deg)  # Переводим в радианы

# Длина векторов
length = 1

# Координаты вектора A (по оси X)
A_x = length
A_y = 0

# Координаты вектора B (под углом angle_deg)
B_x = length * np.cos(angle_rad)
B_y = length * np.sin(angle_rad)

# Создаем график
fig, ax = plt.subplots()

# Рисуем векторы
ax.arrow(0, 0, A_x, A_y, head_width=0.05, head_length=0.1, fc='blue', ec='blue', label='A')
ax.arrow(0, 0, B_x, B_y, head_width=0.05, head_length=0.1, fc='red', ec='red', label='B')

# Отметим текст
plt.text(A_x + 0.1, A_y, 'A', fontsize=12, color='blue')
plt.text(B_x + 0.1, B_y, 'B', fontsize=12, color='red')

# Подпишем угол
plt.text(0.5, 0.1, f"{angle_deg}°", fontsize=14, color='green')

# Настройки осей
ax.set_xlim(-0.2, 1.5)
ax.set_ylim(-0.2, 1.5)
ax.set_aspect('equal')
plt.grid(True)
plt.title('Визуализация угла между векторами A и B')

# Показать график
plt.show()
```

```python
# Разделяем данные на два периода, используя медианную дату (middle_date)
A_values = df[df['Date'] > middle_date]['Rainfall']
B_values = df[df['Date'] <= middle_date]['Rainfall']

# Вычисляем статистические показатели для каждого периода
def calculate_stats(values):
    mean = np.mean(values)
    variance = np.var(values)
    std_dev = np.std(values)
    mean_abs_dev = np.mean(np.abs(values - mean))
    coeff_of_var = (std_dev / mean) * 100  # Умножаем на 100 для процентов
    return mean, variance, std_dev, mean_abs_dev, coeff_of_var

stats_A = calculate_stats(A_values)
stats_B = calculate_stats(B_values)

# Выводим результаты
print("Статистические показатели для периода A (после медианной даты):")  # Уточняем период
print(f"  Среднее значение, мм: {stats_A[0]:.2f}")
print(f"  Дисперсия, мм²: {stats_A[1]:.2f}")  # Добавляем единицы измерения
print(f"  Стандартное отклонение, мм: {stats_A[2]:.2f}")  # Добавляем единицы измерения
print(f"  Среднее отклонение, мм: {stats_A[3]:.2f}")  # Добавляем единицы измерения
print(f"  Коэффициент вариации, %: {stats_A[4]:.2f}")  # Добавляем единицы измерения

print("\nСтатистические показатели для периода B (до или на медианной дате):")  # Уточняем период
print(f"  Среднее значение, мм: {stats_B[0]:.2f}")
print(f"  Дисперсия, мм²: {stats_B[1]:.2f}")  # Добавляем единицы измерения
print(f"  Стандартное отклонение, мм: {stats_B[2]:.2f}")  # Добавляем единицы измерения
print(f"  Среднее отклонение, мм: {stats_B[3]:.2f}")  # Добавляем единицы измерения
print(f"  Коэффициент вариации, %: {stats_B[4]:.2f}")  # Добавляем единицы измерения
```

```python
# Данные для диаграммы
labels = ['Среднее', 'Дисперсия', 'Ст. откл.', 'Ср. откл.', 'Коэф. вар.']
period_A_values = stats_A
period_B_values = stats_B

# Изменяем значения для коэффициента вариации
period_A_values = list(period_A_values)  # Преобразуем в список
period_A_values[4] = period_A_values[4]  # Оставляем как есть

period_B_values = list(period_B_values)
period_B_values[4] = period_B_values[4]  # Оставляем как есть

# Создаем столбчатую диаграмму
x = np.arange(len(labels))  # Положение меток на оси X
width = 0.35  # Ширина столбцов

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, period_A_values, width, label='Период A')
rects2 = ax.bar(x + width/2, period_B_values, width, label='Период B')

# Добавляем подписи, метки и легенду
ax.set_ylabel('Значение')
ax.set_title('Статистические показатели Rainfall для двух периодов')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

# Измененная функция autolabel
def autolabel(rects, is_coeff_of_var=False):
    """Attach a text label above each bar in `rects`,
    displaying its height with limited decimal places.
    Adds '%' sign for coefficient of variation."""
    for i, rect in enumerate(rects):
        height = rect.get_height()
        label = '{:.2f}'.format(height)
        if is_coeff_of_var and i == 4:  # Проверяем, является ли это столбец коэф. вар.
            label += '%'  # Добавляем знак процента
        ax.annotate(label,
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1, is_coeff_of_var=True)  # Указываем, что rects1 содержит коэф. вар.
autolabel(rects2, is_coeff_of_var=True)  # Указываем, что rects2 содержит коэф. вар.

fig.tight_layout()
plt.show()
```

**2. Дисперсия (Variance):**
Дисперсия показывает, насколько сильно различаются данные от среднего значения. Чем выше дисперсия, тем больше различие между отдельными значениями.

Для Периода A: Дисперсия составляет 101.95.

Для Периода B: Дисперсия составляет 94.71.

**Что это означает?**
Дисперсия в Периоде A больше, чем в Периоде B, что говорит о том, что в Периоде A осадки были более изменчивыми, то есть за этот период наблюдались как более высокие, так и более низкие значения осадков.

**3. Стандартное отклонение (Standard Deviation):**
Стандартное отклонение — это корень из дисперсии. Оно также показывает, насколько сильно данные отклоняются от среднего значения, но выражается в тех же единицах измерения (мм в данном случае).

Для Периода A: Стандартное отклонение равно 10.10 мм.

Для Периода B: Стандартное отклонение равно 9.73 мм.

**Что это означает?**
Это также подтверждает, что в Периоде A осадки были более изменчивыми, чем в Периоде B, потому что стандартное отклонение в Периоде A выше, чем в Периоде B. В Периоде B осадки более стабильны.

**4. Среднее отклонение (Mean Deviation):**
Среднее отклонение — это среднее значение абсолютных отклонений от среднего значения. Оно показывает, насколько сильно в среднем значения отклоняются от среднего.

Для Периода A: Среднее отклонение равно 5.12 мм.

Для Периода B: Среднее отклонение равно 4.91 мм.

**Что это означает?**
Среднее отклонение в Периоде A больше, что означает, что значения осадков в этом периоде в среднем отклонялись больше от среднего значения по сравнению с Периодом B. В Периоде B отклонения более сдержаны.

**5. Коэффициент вариации (Coefficient of Variation):**
Коэффициент вариации — это отношение стандартного отклонения к среднему значению, выраженное в процентах. Он показывает степень изменчивости данных относительно среднего значения. Чем выше коэффициент вариации, тем больше изменчивость данных по сравнению с их средним значением.

Для Периода A: Коэффициент вариации составляет 298.48%.

Для Периода B: Коэффициент вариации составляет 294.78%.

```python
# Преобразуем столбец 'Date' в формат datetime
df['Date'] = pd.to_datetime(df['Date'])

# Список годов
years = [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]

# Создаем словарь для хранения данных по годам
rainfall_by_year = {}

# Используем groupby для выделения данных по годам
for year, group in df.groupby(df['Date'].dt.year):
    if year in years:
        rainfall_by_year[year] = group['Rainfall'].reset_index(drop=True)

# Функция для расчета статистических показателей
def calculate_stats(values):
  mean = np.mean(values)
  variance = np.var(values)
  std_dev = np.std(values)
  mean_abs_dev = np.mean(np.abs(values - mean))
  coeff_of_var = (std_dev / mean) * 100  # Умножаем на 100 для процентов
  return mean, variance, std_dev, mean_abs_dev, coeff_of_var

# Выводим статистику для каждого года
for year, data in rainfall_by_year.items():
    stats = calculate_stats(data)
    print(f"Статистические показатели Rainfall для {year} года:")
    print(f"  Среднее значение: {stats[0]:.2f}")
    print(f"  Дисперсия: {stats[1]:.2f}")
    print(f"  Стандартное отклонение: {stats[2]:.2f}")
    print(f"  Среднее отклонение: {stats[3]:.2f}")
    print(f"  Коэффициент вариации: {stats[4]:.2f}%")
    print("\n")
```

```python
# Данные для диаграммы
years = list(rainfall_by_year.keys())
means = [calculate_stats(rainfall_by_year[year])[0] for year in years]
variances = [calculate_stats(rainfall_by_year[year])[1] for year in years]
std_devs = [calculate_stats(rainfall_by_year[year])[2] for year in years]
mean_abs_devs = [calculate_stats(rainfall_by_year[year])[3] for year in years]
coeff_of_vars = [calculate_stats(rainfall_by_year[year])[4] for year in years]

# Собираем данные
data = pd.DataFrame({
    'Среднее значение': means,
    'Дисперсия': variances,
    'Стандартное отклонение': std_devs,
    'Среднее отклонение': mean_abs_devs,
    'Коэффициент вариации': coeff_of_vars
}, index=years)

# Транспонируем таблицу
data = data.T  # Теперь: строки - показатели, столбцы - года

# Параметры для графика
fig, ax = plt.subplots(figsize=(14, 8))

n_years = len(data.columns)  # Кол-во годов
bar_width = 0.8 / n_years  # Ширина одной колонки
index = np.arange(len(data.index))  # Индексы показателей (x-позиция)

# Для каждого года рисуем бары с небольшим сдвигом
for i, year in enumerate(data.columns):
    ax.bar(index + i * bar_width, data[year], width=bar_width, label=str(year))

# Оси и подписи
ax.set_xlabel('Показатели')
ax.set_ylabel('Значение')
ax.set_title('Статистические показатели по годам (каждый показатель отдельно)')
ax.set_xticks(index + bar_width * (n_years - 1) / 2)
ax.set_xticklabels(data.index, rotation=45, ha='right')
ax.legend(title='Год', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
